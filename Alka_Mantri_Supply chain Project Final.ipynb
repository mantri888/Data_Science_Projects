### Encoding the data



One-hot encoding is a method of encoding categorical variables into a set of binary variables. For each categorical variable, one binary variable is created for each category. The binary variable is set to 1 if the observation belongs to that category, and 0 otherwise.

### Converting categorical to dummy variables

```python
# Select only the categorical columns
categorical_columns = scaled_df.select_dtypes(include=['object']).columns

# Perform one-hot encoding
encoded_df = pd.get_dummies(scaled_df, columns=categorical_columns)

# Now, encoded_df contains the original numerical columns along with one-hot encoded categorical variables
```

## Silhouette Analysis

```python
# Silhouette Analysis
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

range_n_clusters = range(2, 11)  # Range of clusters from 2 to 10
for num_clusters in range_n_clusters:
    
    # Initialize KMeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(selected_features_df)
    
    # Get cluster labels
    labels = kmeans.labels_
    
    # Calculate silhouette score
    silhouette_avg = silhouette_score(selected_features_df, labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```
